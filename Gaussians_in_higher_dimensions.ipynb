```
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, IntSlider, Layout
%matplotlib inline
sns.set()
 
size = 1000
x = np.random.randn(size)
plt.figure(figsize=(12,4))
sns.distplot(x)
plt.xlim(-3,3)
plt.ylim(0,0.5)
plt.figure(figsize=(12,1))
plt.scatter(x,[0]*size, alpha=.1, s=200, marker='x')
plt.xlim(-3,3)
plt.ylim(-1,1)
plt.show()
```

```
size = 20000
x = np.random.randn(size)
y = np.random.randn(size)
plt.figure(figsize=(12,4))
sns.distplot(x)
plt.xlim(-3,3)
plt.ylim(0,0.5)
fig, ax = plt.subplots(1, 2, figsize=(18,12), gridspec_kw={'width_ratios': [4, 1.5]})
ax[0].scatter(x, y, alpha=.1, s=200, marker='x')
ax[0].set_xlim(-3,3)
ax[0].set_ylim(-3,3)
sns.distplot(y,vertical=True,ax=ax[1])
plt.ylim(-3,3)
plt.xlim(0,0.5)
plt.show()
```

```
size = 20000
fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(np.random.randn(size), np.random.randn(size), np.random.randn(size), alpha=.1, s=200, marker='x')
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_zlim(-2,2)
plt.show()
```

```
max_dim = 500
x = np.random.randn(50000, max_dim)
n = [np.linalg.norm(x[:,:d], axis=-1) for d in range(max_dim+1)]
@interact(dimension=IntSlider(min=1, max=max_dim, step=1, value=1, layout=Layout(width='50%')))
def f(dimension):
  plt.figure(figsize=(12,8))
  sns.distplot(n[dimension])
  plt.xlabel('Magnitude')
  plt.xlim(0,int(np.sqrt(max_dim))+3)
  plt.xticks(range(int(np.sqrt(max_dim))+4))
  plt.ylim(0,0.8)
```

```
mu = [np.mean(n[d]) for d in range(1,max_dim+1)]
plt.figure(figsize=(12,8))
plt.scatter(range(1,max_dim+1),mu,marker='x')
plt.xlabel('Dimension')
plt.ylabel('Magnitude')
plt.title('Mean of Chi distribution') #https://en.wikipedia.org/wiki/Chi_distribution
plt.show()
```
